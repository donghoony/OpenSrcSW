```python
import csv

```

```python
# 1. 서울의 일교차
"""
서울에서 일교차가 가장 큰 날, 최저기온이 가장 
2021-1-1 ~ 12-31
"""

with open("temperature.csv", 'r', encoding="cp949") as f:
  data = csv.reader(f);
  data = [(x[0], float(x[3]), float(x[4])) for x in list(data)[8:-1]]
  # x : (날짜, 최저, 최고)
  data.sort(key=lambda x: x[1])
  print(f"최저기온이 가장 낮은 날 : {data[0][0]} : {data[0][1]}도")
  data.sort(key=lambda x: -x[2])
  print(f"최고기온이 가장 높은 날 : {data[0][0]} : {data[0][2]}도")
  data.sort(key= lambda x: -(x[2] - x[1]))
  print(f"일교차가 가장 큰 날 : 일교차 {data[0][0]} : {data[0][2] - data[0][1]}도")

```

```python
"""
# 2. 지하철 이용객
2021/12월 각 노선별로 이용객이 많은 3개 역과 가장 적은 3개 역 출력 (노선별로)
"""
from collections import defaultdict
with open("tmoney.csv", 'r', encoding="cp949") as f:
  data = csv.reader(f);
  data = [(x[1], x[3], int(x[4].replace(",", "", -1)) + int(x[5].replace(",", "", -1))) for x in list(data)[1:]]
  d = defaultdict(list)
  for line, name, num in data:
    d[line].append((name, num))
  for line in d:
    d[line].sort(key=lambda x:-x[1])
    print(f"노선 : {line}")
    for i in range(3):
      print(f"최다 이용 역 #{i+1} : {d[line][i][0]}, {d[line][i][1]}명 이용")
    for i in range(3):
      print(f"최저 이용 역 #{i+1} : {d[line][-(i+1)][0]}, {d[line][-(i+1)][1]}명 이용")
```
